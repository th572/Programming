```python
from mpl_toolkits import mplot3d

import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes(projection='3d')

def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 4)
y = np.linspace(-6, 6, 4)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

print(X, Y)
print(Z)
```

```python
c1 = 10
c2 = 0.15
c3 = 0.005
c4 = 0.02

y0 = [1,0,0]
k = 1

def Oregonator (t,Y):
    x = Y[0]
    y = Y[1]
    z = Y[2]
    dxdt = c1 + c2*x - x - x*y**2
    dydt = (x + x*y**2 - y)/c3
    dzdt = (y-z)/c4
    return np.array([dxdt,dydt,dzdt], dtype = float)

t = np.linspace(0, 10, 5000)

Y = odeint(Oregonator, y0, t)

plt.plot(t,Y)
plt.xlabel('t / s')
plt.ylabel('concentration / M')
```
